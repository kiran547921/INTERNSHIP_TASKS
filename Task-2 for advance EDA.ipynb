```python
class Solution(object):
    def defangIPaddr(self, address):
        """
        :type address: str
        :rtype: str
        """
        a=address.replace('.','[.]')
        return a
address=input()
obj=Solution()
print(obj.defangIPaddr(address))
```

```python
class Solution(object):
    def findNumbers(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        c=0
        for i in nums:
            if len(str(i))%2==0:
                c=c+1
        return c
nums=list(map(int,input().split()))
obj=Solution()
print(obj.findNumbers(nums))
```

```python
class Solution(object):
    def numIdenticalPairs(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        freq = {}
        pairs = 0
        for n in nums:
            pairs += freq.get(n, 0)
            freq[n] = freq.get(n, 0) + 1
        return pairs
nums=list(map(int,input().split()))
obj=Solution()
print(obj.numIdenticalPairs(nums))    
```

```python
class Solution(object):
    def smallerNumbersThanCurrent(self, nums):
        a=[]
        for i in range(len(nums)):
            c=0
            for j in range(len(nums)):
                if nums[j]<nums[i]:
                    c=c+1
            a.append(c)   
        return a 
nums=list(map(int,input().split())) 
obj=Solution()
obj.smallerNumbersThanCurrent(nums)
```

```python
class Solution(object):
    def subtractProductAndSum(self, n):
        """
        :type n: int
        :rtype: int
        """
        c1=1
        c2=0
        d=str(n)
        for i in d:
            c1=c1*int(i)
            c2=c2+int(i)
        z=c1-c2
        return z 
n=int(input()) 
obj=Solution()
obj.subtractProductAndSum(n)
```

```python
class Solution(object):
    def xorOperation(self, n, start):
    
        res = 0
        for i in range(n):
            res ^= start + 2 * i
        return res
start=int(input())
n=int(input())
obj=Solution()
obj.xorOperation(n,start)
    
```
